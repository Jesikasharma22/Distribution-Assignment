#1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

Probability of rolling a 3 exactly 5 times (30 rolls)

This follows a binomial distribution where:

- n=30 trials

p= 1/6 (probability of getting a 3)
	​

x=5

P(X=x)= (n x) p^x (1−p)^n−x

Excel formula:
```
=BINOM.DIST(5,30,1/6,FALSE)
```
Answer: ≈ 0.102 (10.2%)

#3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

Given:

- Mean = 50

- Standard deviation = 5

Compute Z-scores:
                  Z= (X−μ)/​σ

For 45 → Z = -1

For 55 → Z = +1          

Answer:

About 68% of values lie between 45 and 55
(Empirical Rule)


#4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

Standardization (Z-score)

Concept:

- Converts data into a common scale

- Mean becomes 0, standard deviation becomes 1

Formula:
```
Z=(X−μ)/σ
```
Importance:

- Compare different datasets
- Detect outliers
- Used in ML, statistics

Effect on data:

- Shape stays same
- Only scale changes

#8. How does log transformation help in stabilizing variance and making data more normally distributed?

Log Transformation

Applies:
log(X)

Benefits:

- Reduces skewness
- Stabilizes variance
- Makes data more normal

Example:

- Original: 10, 100, 1000 (very spread out)
- Log: 1, 2, 3 (more balanced)

Used in:

- Finance
- Machine learning
- Data analysis
